## Step 1: Import Libraries

We import tools for:
- loading the model
- handling text
- computing similarity
- visualization

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Model

We use a pre-trained BERT-based model to convert sentences into embeddings.

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

## Step 3: Input Sentences

We define sample sentences to compare their meanings.

```python
sentences = [
    "I love data",
    "I enjoy coding",
    "The weather is nice today",
    "It is a sunny day",
    "I dislike bugs in code"
]
```

## Step 4: Generate Embeddings

Each sentence is converted into a numerical vector.
These vectors capture meaning.

```python
embeddings = model.encode(sentences)
```

## Step 5: Compute Similarity

We use cosine similarity to compare sentence meanings.
Higher values = more similar.

```python
similarity_matrix = cosine_similarity(embeddings)
```

## Step 6: Visualize Results

We display similarity scores using a heatmap.

```python
plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix,
            annot=True,
            xticklabels=sentences,
            yticklabels=sentences,
            cmap='Blues')

plt.title("Sentence Similarity Matrix")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

## Step 7: Interpretation

- Sentences with similar meaning have high similarity scores
- Unrelated sentences have low scores
- The model captures semantic meaning, not just exact words

## Extra: Compare Two Sentences

We can check similarity between specific pairs.

```python
print(similarity_matrix[0][1])  # "I love data" vs "I enjoy coding"
print(similarity_matrix[2][3])  # weather vs sunny day
```
